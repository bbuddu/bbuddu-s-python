```python
import requests

from bs4 import BeautifulSoup as bs
```

```python
# 주소입력 변수 생성
url = 'https://www.naver.com'
# 네이버에 요청
respnse = requests.get(url)
```

```python
html_data =  respnse.text               #텍스트 형태로 변환시켜 줌
```

```python
# 파싱
# html의 형태로 변환 시켜주겠다~                    #request 에서 받아와서  text로 바꿔주고 다시 BeautifulSoup 에서  html 형태로 파싱작업을 하기 위해서 변형
soup = bs(html_data, 'html.parser')
```

```python
# find() : HTML 안에서 해당 태그의 첫번쨰 정보 출력
# fund_all() : HTML 안에서 해당 태그의 모든 정보를 리스트 형식으로 출력
print(soup.find('a'))
```

```python
print(soup.find_all('a'))
```

```python
print(soup.find(id = 'newsstand'))
```

```python
print(soup.find_all('span',limit=10)[1])     # 리스트 형태에서 대괄호[] 안에 숫자 넣으면 그 순서에 맞는 값만 출력 가능. 
```

```python
# 해당 태그에서 calss 값을 기준으로 출력             # 대괄호[] -> 리스트 형태
print(soup.find_all('span', attrs={                # 중괄호{} -> 딕셔너리 형태
    'class' :  'blind'
}))
```

```python
# 1. 리스트의 값들을 하나씩 출력

# 2. 리스트의 각각text들을 하나씩 출력

# 3. text들을 리스트로 생성
```

```python
# 1. 리스트의 값들을 하나씩 출력
# case1
span_list[0]
span_list[1]
...
# case2
for i in span_list : 
    print(i)
```

```python
# 2. 리스트의 각각text들을 하나씩 출력
for i in span_list : 
    print(i.get_text())
```

```python
# 3. text들을 리스트로 생성
data_list = []
for i in span_list:
    data_list.append(i.get_text())
data_list
```

```python
# 3. text들을 리스트로 생성
# case2
data_list2 = []
for i in range(0,len(span_list),1):
    data = span_list[i].get_text()
    data_list.append(data)

data_list

```

```python
# 3. text들을 리스트로 생성
# case3
# map 함수 : map(함수, 리스트 형태의 데이터)
def change(x):
    #print(x.get_text())
    return x.get_text()

list(map(change, span_list))
```

```python
# case4
# map(lambda함수, 리스트의 형태 데이터)

list(map(lambda x :x.get_text() , span_list))
```
