### 예제 문제
1. csv 폴더에 있는 sales records.csv 파일 로드
2. 'Order ID', 'Total Revenue', 'Total Cost', 'Total profit' 컬럼을 삭제
3. 컬럼의 이름을 변경 ['대륙', '국가', '타입', '채널', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']
4. 대륙, 국가를 기준으로 오름차순 정렬
5. 인덱스를 초기화(기존의 인덱스는 삭제)

```python
import pandas as pd
```

```python
salesrecords =pd.read_csv("../csv/Sales Records.csv")
```

```python
salesrecords.head()
```

```python
salesrecords.drop(['Order ID','Total Revenue','Total Cost','Total Profit'], axis=1, inplace=True)
```

```python
salesrecords.columns=['대륙', '국가', '타입', '채널', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']
```

```python
salesrecords.head()
```

```python
salesrecords.sort_values(['대륙','국가'], inplace=True)
```

```python
salesrecords
```

```python
salesrecords.reset_index(drop=True, inplace=True)
```

```python
salesrecords.head(1)
```

```python
## 총 단가, 총 원가, 총 이윤

## 총 단가 = 단가 * 판매단위

salesrecords['총 단가']=salesrecords['단가'] * salesrecords['판매단위']
salesrecords.head(2)
```

```python
## 총 원가 = 원가 * 판매단위
salesrecords['총 원가'] = salesrecords['원가'] * salesrecords['판매단위']
```

```python
salesrecords
```

```python
## 총 이윤 = 총 단가 - 총 원가
salesrecords['총 이윤'] = salesrecords['총 단가'] - salesrecords['총 원가']

## 총 이윤 = 판매단위 * (단가 - 원가)
salesrecords['총 이윤2'] = salesrecords['판매단위'] * (salesrecords['단가'] - salesrecords['원가'])

```

```python
salesrecords.head(2)
```

### loc[ ]

    - loc[행의 위치, 열의 위치]

### iloc[ ]
    - 특정 행이나 특정 열을 추출할 때 사용
    - iloc[행의 위치 인덱스 값, 열의 위치 인덱스 값]

### 그룹화
 - groupby(그룹화 기준이 되는 컬럼)

```python
salesrecords.groupby("국가").sum()['총 이윤']
```

```python
## 가장 많은 이윤을 내는 대륙은 어디인가?

salesrecords.groupby('대륙').sum('총 이윤')[['총 이윤']]

## 그룹화 - 합계 - 오름차순 or 내림차순 - 하위 값 or 상위 값
```

```python
test = salesrecords.groupby('대륙').sum()[['총 이윤']]

## 내림차순 정렬 -> head값 출력
test.sort_values(['총 이윤'], ascending=False).head(1)
```

```python
## 국가별 총 이윤이 높은 상위 10개의 국가를 울력
## 국가별 그룹화
test2 = salesrecords.groupby('국가').sum()[['총 이윤']]
## 총 이윤을 기준으로 내림차순 정렬
group_df = test2.sort_values(['총 이윤'], ascending=False)
## 상위 10개
top10 = group_df.head(10)
## 하위 10개
bottom10 = group_df.tail(10)
```
